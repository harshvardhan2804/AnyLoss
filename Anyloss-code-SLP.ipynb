# **SLP WITH MSE - BASELINE MODEL 1**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

data = pd.read_csv('/content/drive/MyDrive/Datasets/breast_cancer.csv')
data.head()
```

```python
Y = data['diagnosis']
Y = data['diagnosis'].replace({'M': 1, 'B': 0})
X_features = ['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave points_mean','symmetry_mean','fractal_dimension_mean']
X = data[X_features]
```

```python
X_train_a, X_test_a, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score

def compute_linear(features, weights):
    return np.dot(features, weights.T)

def compute_loss(predictions, target):
    return ((target - predictions) ** 2).sum() / 2.0

def compute_gradient(features, predictions, target):
    return -np.dot((target - predictions), features)

def update_weights(weights, alpha, grad):
    return weights - (alpha * grad)

def classify(features, weights):
    predictions = compute_linear(features, weights)
    return np.array([1 if predictions[i] > 0 else -1 for i in range(features.shape[0])])

weights = np.zeros(X_train_a.shape[1])
iterations = 800
errors = []
alpha = 0.05

for _ in range(iterations):
    predictions = compute_linear(X_train_a, weights)
    errors.append(compute_loss(predictions, y_train))
    grad = compute_gradient(X_train_a, predictions, y_train)
    weights = update_weights(weights, alpha, grad)

final_predictions_test = classify(X_test_a, weights)
```

# **SLP WITH BINARY CROSS ENTROPY - BASELINE MODEL 2**

```python
import numpy as np

def activation_sigmoid(features, weights):
    linear_combination = np.dot(features, weights.T).astype(float)
    return 1.0 / (1 + np.exp(-linear_combination))

def compute_cost(probabilities, labels):
    return ((-labels * np.log(probabilities)) - ((1 - labels) * np.log(1 - probabilities))).mean()

def compute_gradient(features, probabilities, labels, sample_size):
    return np.dot(features.T, (probabilities - labels)) / sample_size

def update_weights(weights, alpha, grad):
    return weights - (alpha * grad)

def classify_samples(features, weights, decision_boundary):
    probabilities = activation_sigmoid(features, weights)
    return [0 if probabilities[i] <= decision_boundary else 1 for i in range(features.shape[0])]

weights = np.zeros(X_train_a.shape[1])
decision_boundary = 0.5
iterations = 1000
error_log = []
alpha = 0.1
sample_size = y_train.shape[0]

for _ in range(iterations):
    probabilities = activation_sigmoid(X_train_a, weights)
    error_log.append(compute_cost(probabilities, y_train))
    grad = compute_gradient(X_train_a, probabilities, y_train, sample_size)
    weights = update_weights(weights, alpha, grad)

final_predictions = classify_samples(X_train_a, weights, decision_boundary)
```

# **SLP WITH ANYLOSS**

```python
import numpy as np

def activation_function(features, weights):
    linear_output = np.dot(features, weights.T).astype(float)
    return 1.0 / (1 + np.exp(-linear_output))

# Approximation Function
def approximate_activation(probabilities):
    scaled_values = (scaling_factor * (probabilities - decision_threshold)).astype(float)
    return 1.0 / (1 + np.exp(-scaled_values))

#Loss Functions

# F-beta Score

def compute_fbeta(beta, sum_yhat, sum_yhat_y, beta_sum_y):
    return 1 - ((1 + beta) * sum_yhat_y / (beta_sum_y + sum_yhat))

def gradient_fbeta(prob, features, yhat, labels, beta, sum_yhat, sum_yhat_y, beta_sum_y):
    term = scaling_factor * yhat * (1 - yhat) * prob * (1 - prob)
    return -((1 + beta) * (np.dot(labels * term * (beta_sum_y + sum_yhat), features) - np.dot(term * sum_yhat_y, features))) / ((beta_sum_y + sum_yhat) ** 2)

# Accuracy Score
def compute_accuracy(total_y, sum_yhat, sum_yhat_y, total_samples):
    return 1 - ((total_samples - total_y - sum_yhat + 2 * sum_yhat_y) / total_samples)

def gradient_accuracy(prob, features, yhat, labels, total_samples):
    term = scaling_factor * yhat * (1 - yhat) * prob * (1 - prob)
    return (np.dot(term, features) - 2 * np.dot(labels * term, features)) / total_samples

# Geometric Mean
def compute_gmean(total_y, sum_yhat, sum_yhat_y, total_samples):
    return 1 - ((sum_yhat_y * (total_samples - sum_yhat - total_y + sum_yhat_y) / (total_y * (total_samples - total_y))) ** 0.5)

def gradient_gmean(prob, features, yhat, labels, total_y, sum_yhat, sum_yhat_y, total_samples):
    term = scaling_factor * yhat * (1 - yhat) * prob * (1 - prob)
    repeated_1 = np.dot(labels * term, features)
    repeated_2 = total_samples - sum_yhat - total_y + sum_yhat_y
    return -2 * ((repeated_1 * repeated_2) + (-np.dot(term, features) + repeated_1) * sum_yhat_y) / (total_y * (total_samples - total_y) * sum_yhat_y * repeated_2) ** 0.5

# Balanced Accuracy
def compute_balanced_accuracy(total_y, sum_yhat, sum_yhat_y, total_samples):
    return 1 - ((total_samples * (sum_yhat_y + total_y) - total_y * (sum_yhat + total_y)) / (2 * total_y * (total_samples - total_y)))

def gradient_balanced_accuracy(prob, features, yhat, labels, total_y, total_samples):
    term = scaling_factor * yhat * (1 - yhat) * prob * (1 - prob)
    return -(total_samples * np.dot(labels * term, features) - total_y * np.dot(term, features)) / (2 * total_y * (total_samples - total_y))

def update_weights(weights, learning_rate, grad):
    return weights - (learning_rate * grad)

def classify(features, weights):
    probabilities = activation_function(features, weights)
    return [1 if val > decision_threshold else 0 for val in probabilities]

weights = np.zeros(X_train_a.shape[1])
decision_threshold = 0.5
iterations = 1000
scaling_factor = 73
error_log = []
learning_rate = 0.1
total_y = np.sum(y_train)
total_samples = y_train.shape[0]

for _ in range(iterations):
    prob = activation_function(X_train_a, weights)
    yhat = approximate_activation(prob)
    sum_yhat = np.sum(yhat)
    sum_yhat_y = np.dot(yhat, y_train)

    loss = compute_accuracy(total_y, sum_yhat, sum_yhat_y, total_samples)
    error_log.append(loss)

    gradient = gradient_accuracy(prob, X_train_a, yhat, y_train, total_samples)

    weights = update_weights(weights, learning_rate, gradient)

final_predictions = classify(X_train_a, weights)
```

```python
# Predictions on test set
test_predictions = classify(X_test_a, weights)

def compute_test_accuracy(y_test, test_predictions):
    correct = np.sum(y_test == test_predictions)
    return correct / len(y_test)

accuracy = compute_test_accuracy(y_test, test_predictions)
print(f"Test Accuracy using Anyloss: {accuracy:.4f}")
```
